```python
# cleaning

# Q. DOB object to datetime format
df['dob'] = pd.to_datetime(df['dob'])
print(df.info())

# Q. Number of null/none/N.A entries, if NA entries are in salary fill them with median, else in any other column remove that entry/row
print("Total NA entries: ",df.isna().sum())
median_salary = df['salary'].median()
df['salary'] = df['salary'].fillna(median_salary) # df['salary'].fillna(median_salary,inplace=True)
df.dropna(subset = [col for col in df.columns if col!='salary'],inplace=True)
print("Total NA entries: ",df.isna().sum().sum())

# Q. Find duplicates/false entries using name and age, print them and then remove them.
duplicate_count = df.duplicated(subset=['name', 'age']).sum()
print("Duplicate entries:", duplicate_count)
duplicates = df[df.duplicated(subset=['name', 'age'], keep=False)]
print("Duplicate entries based on 'name' and 'age':\n", duplicates)
df = df.drop_duplicates(subset=['name', 'age'], keep='first')
duplicate_count = df.duplicated(subset=['name', 'age']).sum()
print("Duplicate entries:", duplicate_count)

# Q. Find dob below year 1985, and change their salary by increasing 25%
df_before_1985 = df[df['dob'].dt.year < 1985]
print(df_before_1985)
df.loc[df['dob'].dt.year < 1985,'salary'] *= 1.25 # label based indexing

# Q. Drop a column
df = df.drop(columns=['color'])
df

# Q. Replace all entries before 1985 to "1990-01-01"
print(df_before_1985)
df.loc[df['dob'].dt.year < 1985, 'dob'] = pd.to_datetime("1990-01-01")
print(df_before_1985)

# Q. Rename Column sex to gender
df.rename(columns = {'sex':'gender'},inplace=True)
df
```

```python
department_data = {
    'DepartmentName': ['HR', 'Finance', 'Engineering', 'Marketing'],
    'id' : [1,2,3,4]
}

departments = pd.DataFrame(department_data)
departments
```

```python
# data transformations (mergining and joining) & statistical functions

# Q. create a column with name Key = 'name-gender(f/m)-age'
df['gender'] = df['gender'].astype(str)
keys = df['name']+'-'+df['gender'].str[0]+'-'+df['age'].astype(str) # df['key'] instead of keys
df = pd.concat([df,keys],axis=1) # df['key'] = keys
df.rename(columns = {'key':'keys'},inplace=True)
df

# Q. join df and departments on id, if id>4, perform mod%4, and merge with resultant id
df['id'] = (df.index%4) + 1
print(df.merge(departments,left_on='id',right_on='id',how='inner')) # df.merge(df, how='inner' 'left' 'right' , on='')

# ** merge vs join - merge connect df on columns, join connects on indexes
# df.join(newdf, how='left' 'inner' 'right', on='')

# sort on basis of age and then salary
print(df.sort_values(by='dob',ascending=False)) # descending
print(df.sort_values(by='salary',ascending=True))
print(df.sort_index())

# Apply function - where salary > 50k is increased by 10% and if less than that increased by 20%
def SALARY(sal):
  if sal>50000:
    return sal*1.1
  else:
    return sal*1.2

df['salary'] = df['salary'].map(SALARY)
df

# find maxima and minima in salary
print(df['salary'].min())
print(df['salary'].max())

# find top 1% earning employees
# ** top 1% means 99th percentile of salary column
sal99 = df['salary'].quantile(0.99)
result_df = df[df['salary']>=sal99]
result_df = result_df[['name','salary','id']]
result_df_sal = result_df['salary'].tolist()
print(result_df)
print(result_df_sal)

# find and print employee with salary greater than average
avgsal = df['salary'].mean() # .median()
print(avgsal)
result_df = df[df['salary']>avgsal]
print(result_df)
```
